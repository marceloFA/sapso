```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 13})
```

```python
# Importing data
data = pd.read_csv('parallel_sapso_benchmark.txt')
print(data.info())
data.head()
print('\nThose were the tested functions:\n',data['f_name'].unique())
```

```python
plt.figure(figsize=(8,4))
data[data['f_name'] == 'alpine'].groupby('n_dims')['mean_min'].plot()
plt.ylim(-1e-5,1e-5)
plt.legend(title='Number of dimensions', loc='upper right')
plt.title("Results of optmization for the Alpine function")
plt.ylabel("Minimum Found")
plt.xlabel("Execution")
```

```python
plt.figure(figsize=(8,4))
data[data['f_name'] == 'ellipsoid'].groupby('n_dims')['mean_min'].plot()
plt.ylim(0,5)
plt.legend(title='Number of dimensions', loc='upper right')
plt.title("Results of optmization for the Ellipsoid function")
plt.ylabel("Minimum Found")
plt.xlabel("Execution")
```

```python
print('Mean minimum was:')
data.groupby('f_name')['mean_min'].mean()
```
