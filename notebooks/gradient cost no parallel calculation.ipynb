```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 13})
```

```python
df = pd.read_csv('grad_cost_no_parallel_calculation.csv')
df = df.drop(columns=["Unnamed: 4"])
df
```

```python
plt.bar(range(10),height=df['%'])
```

```python
labels = 'Gradient Calculation', 'All other tasks'
mean_percentage = df['%'].mean()
rest = 100. - mean_percentage
sizes = [mean_percentage,rest]
explode = (.2, 0)
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.set_title("Mean percentage of execution time consumption discriminated task")
ax1.pie(sizes,explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=True, startangle=15)
```
